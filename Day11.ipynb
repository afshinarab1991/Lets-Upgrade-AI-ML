```python
data_nums_cor = data_nums.drop(["StandardHours","EmployeeCount"], axis = 1)
cor_tab = data_nums_cor.corr()
cor_tab
```

```python
mask = np.triu(np.ones_like(cor_tab, dtype=np.bool))
f, ax = plt.subplots(figsize=(10, 10))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(cor_tab, mask=mask, cmap=cmap, vmax=.79,vmin= -.2, center=0.1,
            square=True, linewidths=.4, cbar_kws={"shrink": .4});
```

```python
# generating a list of correlation to find out the highest correlations:
cor_shape = list(range(int(cor_tab.shape[0])))
tlist = []
for x in cor_shape: #filling the diagonal values with zeros
    for x in cor_shape:
        cor_tab.iloc[x,x] = 0
for x in cor_shape: #generating a list with all the correlation values
    for y in cor_shape:
        point = cor_tab.iloc[x,y] 
        if point not in tlist:
            tlist.append(point)
        else:
            continue
    tlist = sorted(tlist, reverse= True)

hcp = tlist[0:4] #highest correlations (positive)
hcn = tlist[-4:] #highest correlations (negative)
```

```python
def getIndexes(dfObj, value):
    ''' Get index positions of value in dataframe i.e. dfObj.'''
    listOfPos = list()
    # Get bool dataframe with True at positions where the given value exists
    result = dfObj.isin([value])
    # Get list of columns that contains the value
    seriesObj = result.any()
    columnNames = list(seriesObj[seriesObj == True].index)
    # Iterate over list of columns and fetch the rows indexes where value exists
    for col in columnNames:
        rows = list(result[col][result[col] == True].index)
        for row in rows:
            listOfPos.append((row, col, value))
    # Return a list of tuples indicating the positions of value in the dataframe
    return listOfPos
```

```python
cords_p = []
for i in hcp: #highest positive correlations with their respective name of the row/column
    g = getIndexes(cor_tab,i)
    cords_p.append(g)
pc = [[cords_p[0:4][0][0]],[cords_p[0:4][1][0]],[cords_p[0:4][2][0]],[cords_p[0:4][3][0]]]
pc
```

```python
cords_n = []
for i in hcn: #highest negative correlations with their respective name of the row/column
    g = getIndexes(cor_tab,i)
    cords_n.append(g)
nc = [[cords_n[0:4][0][0]],[cords_n[0:4][1][0]],[cords_n[0:4][2][0]],[cords_n[0:4][3][0]]]
nc
```
