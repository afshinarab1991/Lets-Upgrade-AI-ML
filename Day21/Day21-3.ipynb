```python
import pandas as pd
import statsmodels.api as sm
```

```python
data = pd.read_csv("general_data.csv")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data = data.fillna(method='bfill')
data = data.dropna()
data.isna().sum()
```

```python
y = data[["Attrition"]]
y.head()
```

```python
x = data.drop(["Attrition"],axis =1)
```

```python
x.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for i in x.columns:
    if x[i].dtypes == "O":
        x[i] = le.fit_transform(x[i])
    else:
        pass
```

```python
y["Attrition"] = le.fit_transform(y["Attrition"])
```

```python
y.head()
```

```python
x.head()
```

```python
x.corr()
```

```python
x_constant = sm.add_constant(x)
x_constant = x_constant.drop(["Over18","StandardHours","EmployeeCount"],axis = 1)
```

```python
x_constant.shape
```

```python
y.shape
```

```python
Logistic = sm.Logit(y,x_constant)
```

```python
result = Logistic.fit()
```

```python
result.summary2()
```
